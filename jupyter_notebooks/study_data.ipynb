# Plot Data

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(mf_data['steps'], mf_data['proportion pellet'], label="mean field")
```

```python
# proportion pellet
plt.plot(og_data['steps'], og_data['proportion pellet'], label="original")
plt.plot(mk_data['steps'], mk_data['proportion pellet'], label="markov")
plt.plot(mf_data['steps'], mf_data['proportion pellet'], label="mean field")
plt.legend()
```

```python
# proportion floor
plt.plot(og_data['steps'], og_data['proportion floor'], label="original")
plt.plot(mk_data['steps'], mk_data['proportion floor'], label="markov")
plt.plot(mf_data['steps'], mf_data['proportion floor'], label="mean field")
plt.legend()
```

```python
# proportion floor
plt.plot(og_data['steps'], og_data['volume'], label="original")
plt.plot(mk_data['steps'], mk_data['volume'], label="markov")
plt.plot(mf_data['steps'], mf_data['volume'], label="mean field")
plt.legend()
```
