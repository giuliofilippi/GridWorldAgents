# Plot Data

```python
import matplotlib.pyplot as plt
```

```python
# proportion pellet
plt.plot(og_data['steps'], og_data['proportion_pellet'], label="original")
plt.plot(mk_data['steps'], mk_data['proportion_pellet'], label="markov")
plt.plot(mf_data['steps'], mf_data['proportion_pellet'], label="mean field")
plt.legend()
```

```python
# pickup rate
plt.plot(og_data['steps'], og_data['pickup_rate'], label="original")
plt.plot(mk_data['steps'], mk_data['pickup_rate'], label="markov")
plt.plot(mf_data['steps'], mf_data['pickup_rate'], label="mean field")
plt.legend()
```

```python
# drop rate
plt.plot(og_data['steps'], og_data['drop_rate'], label="original")
plt.plot(mk_data['steps'], mk_data['drop_rate'], label="markov")
plt.plot(mf_data['steps'], mf_data['drop_rate'], label="mean field")
plt.legend()
```

```python
# proportion floor
plt.plot(og_data['steps'], og_data['volume'], label="original")
plt.plot(mk_data['steps'], mk_data['volume'], label="markov")
plt.plot(mf_data['steps'], mf_data['volume'], label="mean field")
plt.legend()
```
